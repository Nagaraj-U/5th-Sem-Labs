```python
ax = sns.countplot(x='gender',hue='test preparation course',palette='Set1',data=df)
ax.set(title='number of students who complete test prep course',xlabel='gender',ylabel='count')
plt.show()
```

```python
plt.figure(figsize=(12,6))
ax = sns.countplot(x='race/ethnicity',hue='gender',palette='Set2',data=df)
ax.set(title='number of students who belong to same group')
plt.show()
```
