```python
def CF(c):
    return (c*(9/5) + 32)

def FC(f):
    return ((f-32)*(5/9))

def CK(c):
    return (c+273.5)

def KC(k):
    return (k-273.5)

def KF(k):
    return (CF(KC(k)))

def FK(f):
    return (CK(FC(f)))

op1 = []
op2 = []
op3 = []
op4 = []
op5 = []
op6 = []

select = "y"

while(select == "y" or select == "Y"):
    ch = int(input("1 : C->F 2 : F->C 3 : C->K 4: K->C 5 : K->F 6 : F->K"))
    if(ch == 1):
        c = int(input("enter celsius temperature"))
        res = CF(c)
        print("C->F is : " , res)
        op1.append(res)
    elif(ch == 2):
        f = int(input("enter fahrenheit temperature"))
        res = FC(f)
        print("F->C is : " , res)
        op2.append(res)
    elif(ch == 3):
        c = int(input("enter celsius temperature"))
        res = CK(c)
        print("C->K is : " , res)
        op3.append(res)
    elif(ch == 4):
        k = int(input("enter kelvin temperature"))
        res = KC(k)
        print("K->C is : " , res)
        op4.append(res)
    elif(ch == 5):
        k = int(input("enter kelvin temperature"))
        res = CF(k)
        print("K->F is : " , res)
        op5.append(res)
    elif(ch == 6):
        f = int(input("enter fahrenheit temperature"))
        res = FK(f)
        print("F->K is : " , res)
        op6.append(res)
        
    select = input("enter Y to continue N to exit")
```

```python
class SentenceReverser:
    sentence = ""
    reverse = ""
    vowelCount = 0
    vowels = ['a','e','i','o','u']
    def __init__(self,sentence):
        self.sentence = sentence
        self.reverseSentence()
        
    def reverseSentence(self):
        self.reverse = " ".join(reversed(self.sentence.split()))
    
    def getVowelCount(self):
        self.vowelCount = sum(s in self.vowels for s in self.sentence.lower())
        return self.vowelCount
    
    def getReverseSentence(self):
        return self.reverse
    
    
items = []
for i in range(3):
    sentence = input("enter phrase ")
    reverser = SentenceReverser(sentence)
    items.append(reverser)
    
sortedItems = sorted(items,key=lambda item:item.getVowelCount(),reverse=True)

print('sorted sentences according to vowel count are ')
for i in range(3):
    print(items[i].getReverseSentence() , ' vowel count : ' ,items[i].getVowelCount())
        
```
